```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from IPython.display import IFrame
```

# Einlesen der Daten

```python
# CSV einlesen

Daten = pd.read_csv("https://raw.githubusercontent.com/martin-wegmann/Modul_Datenvisualisierung/refs/heads/main/Beispieldaten/Beispiel_Iris.csv")
```

```python
# Ueberblick Daten
Daten
```

# Statischen Plot erstellen

```python
# Scatterplot herstellen
## Groesse der Abbildung bestimmen
plt.figure(figsize=(10, 7),dpi=300)
## Daten abbilden
for species, group in Daten.groupby("Spezies"):
    plt.scatter(
        group["Bluetenblatt_Laenge"],
        group["Bluetenblatt_Breite"],
        label=species,
        alpha=1
    )

plt.xlabel("BlÃ¼tenblatt LÃ¤nge")
plt.ylabel("BlÃ¼tenblatt Breite")
plt.title("Iris Datensatz: BlÃ¼tenblatt LÃ¤nge versus Breite")
plt.legend()
plt.tight_layout()
plt.show()
```

# Interactiven Plot erstellen

```python
# Scatterplot herstellen
fig = px.scatter(
    Daten,
    x="Bluetenblatt_Laenge",
    y="Bluetenblatt_Breite",
    color="Spezies",
    title="Iris Datensatz: Kelchblatt LÃ¤nge versus Breite",
    labels={
        "Bluetenblatt_Laenge": "BlÃ¼tenblatt LÃ¤nge",
        "Bluetenblatt_Breite": "BlÃ¼tenblatt Breite"
    }
)
# Abbildung als html abspeichern
fig.write_html("iris_plot.html")
```

```python
# Interaktive Abbildung in Notebook sichtbar machen
IFrame("iris_plot.html", width=700, height=500)
```
