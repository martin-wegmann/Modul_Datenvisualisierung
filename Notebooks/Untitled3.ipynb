```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
Daten = pd.read_csv("https://raw.githubusercontent.com/martin-wegmann/Modul_Datenvisualisierung/refs/heads/main/Beispieldaten/Beispiel_Iris.csv")
```

```python
# Ueberblick Daten
Daten
```

```python
# Scatterplot herstellen
## Groesse der Abbildung bestimmen
plt.figure(figsize=(10, 7),dpi=300)
## Daten abbilden
for species, group in Daten.groupby("Spezies"):
    plt.scatter(
        group["Bluetenblatt_Laenge"],
        group["Bluetenblatt_Breite"],
        label=species,
        alpha=1
    )

plt.xlabel("Blütenblatt Länge")
plt.ylabel("Blütenblatt Breite")
plt.title("Iris Datensatz: Blütenblatt Länge versus Breite")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(
    Daten,
    x="Bluetenblatt_Laenge",
    y="Bluetenblatt_Breite",
    color="Spezies",
    title="Iris Datensatz: Kelchblatt Länge versus Breite",
    labels={
        "Bluetenblatt_Laenge": "Blütenblatt Länge",
        "Bluetenblatt_Breite": "Blütenblatt Breite"
    },
    width=1000,   # width in pixels
    height=800   # height in pixels
)

fig.show()
```
